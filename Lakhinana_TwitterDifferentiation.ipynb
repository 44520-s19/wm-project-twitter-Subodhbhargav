```python
import pickle
import tweepy
from Assignment_keys import consumer_key, consumer_secret, access_token, access_secret

auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_secret)


api = tweepy.API(auth)

england_tweets = api.user_timeline('englandcricket')
aus_tweets = api.user_timeline('cricketcomau')
india_tweets = api.user_timeline('BCCI')
```

# Lexical Diversity

```python
englandtweet = england_tweets[0]
print(englandtweet.text)
austweet = aus_tweets[0]
print(austweet.text)
indiatweet = india_tweets[0]
print(indiatweet.text)
```

```python
englandtokens = englandtweet.text.split()
print(englandtokens)
austokens = austweet.text.split()
print(austokens)
indiatokens = indiatweet.text.split()
print(indiatokens)
```

# Length

```python
print(len(englandtokens))
print(len(austokens))
print(len(indiatokens))
```

```python
lower_tokens = [s.lower() for s in englandtokens]

print(lower_tokens)
unique_lower = set(lower_tokens)
print(unique_lower)
print('Lexical Diversity', len(unique_lower)/len(lower_tokens))


lower_tokens = [s.lower() for s in austokens]

print(lower_tokens)
unique_lower = set(lower_tokens)
print(unique_lower)
print('Lexical Diversity', len(unique_lower)/len(lower_tokens))



lower_tokens = [s.lower() for s in indiatokens]

print(lower_tokens)
unique_lower = set(lower_tokens)
print(unique_lower)
print('Lexical Diversity', len(unique_lower)/len(lower_tokens))
```

```python


def lexical_diversity(text):
    tokens = text.lower().split()
    return len(set(tokens))/len(tokens)
```

```python

print(lexical_diversity(englandtweet.text))
```

```python

print(lexical_diversity(austweet.text))
```

```python
print(lexical_diversity(indiatweet.text))
```
