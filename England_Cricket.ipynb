```python
import pickle
import tweepy
from Assignment_keys import consumer_key, consumer_secret, access_token, access_secret

auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_secret)


api = tweepy.API(auth)

england_tweets = api.user_timeline('englandcricket')
```

# Lexical Diversity

```python
englandtweet = england_tweets[0]
print(englandtweet.text)
```

```python
englandtokens = englandtweet.text.split()
print(englandtokens)
```

# Length

```python
print(len(englandtokens))
```

```python
lower_tokens = [s.lower() for s in englandtokens]

print(lower_tokens)
unique_lower = set(lower_tokens)
print(unique_lower)
print('Lexical Diversity', len(unique_lower)/len(lower_tokens))
```

```python
def lexical_diversity(text):
    tokens = text.lower().split()
    return len(set(tokens))/len(tokens)
```
