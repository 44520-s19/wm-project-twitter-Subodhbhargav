# Lexical Diversity

```python
indiatweet = india_tweets[0]
print(indiatweet.text)
```

```python
indiatokens = indiatweet.text.split()
print(indiatokens)
```

```python
print(len(indiatokens))
```

```python
lower_tokens = [s.lower() for s in indiatokens]

print(lower_tokens)
unique_lower = set(lower_tokens)
print(unique_lower)
print('Lexical Diversity', len(unique_lower)/len(lower_tokens))
```

```python
def lexical_diversity(text):
    tokens = text.lower().split()
    return len(set(tokens))/len(tokens)
```

```python
print(lexical_diversity(indiatweet.text))
```
