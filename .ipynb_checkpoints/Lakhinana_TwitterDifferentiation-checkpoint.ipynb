```python
import pickle
import tweepy
from Assignment_keys import consumer_key, consumer_secret, access_token, access_secret

auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_secret)


api = tweepy.API(auth)

england_tweets = api.user_timeline('englandcricket')
aus_tweets = api.user_timeline('cricketcomau')
india_tweets = api.user_timeline('BCCI')
```

# Lexical Diversity

```python
englandtweet = england_tweets[0]
print(englandtweet.text)
austweet = aus_tweets[0]
print(austweet.text)
indiatweet = india_tweets[0]
print(indiatweet.text)
```

```python
englandtokens = englandtweet.text.split()
print(englandtokens)
austokens = austweet.text.split()
print(austokens)
indiatokens = indiatweet.text.split()
print(indiatokens)
```

# Length

```python
print(len(englandtokens))
print(len(austokens))
print(len(indiatokens))
```

```python
lower_tokens = [s.lower() for s in englandtokens]

print(lower_tokens)
unique_lower = set(lower_tokens)
print(unique_lower)
print('Lexical Diversity', len(unique_lower)/len(lower_tokens))


lower_tokens = [s.lower() for s in austokens]

print(lower_tokens)
unique_lower = set(lower_tokens)
print(unique_lower)
print('Lexical Diversity', len(unique_lower)/len(lower_tokens))



lower_tokens = [s.lower() for s in indiatokens]

print(lower_tokens)
unique_lower = set(lower_tokens)
print(unique_lower)
print('Lexical Diversity', len(unique_lower)/len(lower_tokens))
```

```python


def lexical_diversity(text):
    tokens = text.lower().split()
    return len(set(tokens))/len(tokens)
```

```python

print(lexical_diversity(englandtweet.text))
```

```python

print(lexical_diversity(austweet.text))
```

```python
print(lexical_diversity(indiatweet.text))
```

# Commonly used words

```python
from collections import Counter
total_england = ""
for tweet in england_tweets:
    total_england = total_england+tweet.text 
split_it_england = total_england.split();
englandCounter = Counter(split_it_england);
print(englandCounter)
```

```python
from collections import Counter
total_india = ""
for tweet in india_tweets:
    total_india = total_india+tweet.text 
split_it_india = total_india.split();
indiaCounter = Counter(split_it_india);
print(indiaCounter)
```

```python
from collections import Counter
total = ""
for tweet in aus_tweets:
    total = total+tweet.text 
split_it_aus = total.split();
ausCounter = Counter(split_it_aus);
print(ausCounter)
```

# Sentiment Analysis

```python
import pickle
from nltk_helpers import get_sentiments
print(get_sentiments(englandtweet.text))
print(get_sentiments(indiatweet.text))
print(get_sentiments(austweet.text))
```

```python
from mpl_toolkits.mplot3d import Axes3D

import matplotlib.pyplot as plt
 
# Create a figure

fig = plt.figure()

ax = fig.add_subplot(111, projection='3d')

sentiments = [get_sentiments(tweet.text) for tweet in england_tweets]

xs = []

for sentiment in sentiments:

    xs.append(sentiment['neg'])

ys = [sentiment['neu'] for sentiment in sentiments]

zs = list(map(lambda x: x['pos'],sentiments))

ax.scatter(xs, ys, zs)

from nltk_helpers import split_sentiments


arx,ary,arz=split_sentiments([get_sentiments(tweet.text) for tweet in india_tweets])

ax.scatter(arx, ary, arz,color='r',marker='^')

brx,bry,brz=split_sentiments([get_sentiments(tweet.text) for tweet in aus_tweets])

ax.scatter(brx, bry, brz,color='g',marker='^')

ax.set_xlabel('Negative')

ax.set_ylabel('Neutral')

ax.set_zlabel('Positive')






plt.show()
```

```python
plt.scatter(xs, ys, label='England')
plt.scatter(arx, ary, color='r', marker='s', label='India')
plt.scatter(brx, bry, color='g', marker='s', label='Aus')
plt.xlabel('negative')
plt.ylabel('positive')
plt.title('positive vs negative Sentiments')
plt.legend(loc='upper right', shadow=True)
plt.show()
```
